```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import ds_toolbox.modelling.time_series_utils as ts
import ds_toolbox.modelling.misc_utils as ds
from ds_toolbox.visualisation import corrplot, bokeh_chart, bokeh_candlestick, plot_colors

from bokeh.plotting import show, output_notebook
output_notebook()

pd.set_option("display.max_columns", None)
```

```python
battlegrounds = ['Arizona',
 'Colorado',
 'Florida',
 'Georgia',
 'Iowa',
 'Maine',
 'Michigan',
 'Minnesota',
 'Nevada',
 'New Hampshire',
 'New Mexico',
 'North Carolina',
 'Ohio',
 'Pennsylvania',
 'Texas',
 'Virginia',
 'Wisconsin']

party_colors = ["red", "dodgerblue"]
```

```python
for state in battlegrounds:
    df_2016 = pd.read_csv(f"data/processed/2016_polls/{state}.csv", index_col=0, parse_dates=True)
    df_2020 = pd.read_csv(f"data/processed/2020_polls/{state}.csv", index_col=0, parse_dates=True)
    
    oct2016 = df_2016.loc["2016-10"].iloc[-1][["trump_ave", "clinton_ave", "rep_result","dem_result"]]
    oct2020 = df_2020[["Trump_ave", "Biden_ave"]].iloc[-1]
    
    print(state)
    print("Oct 31 2016")
    print(round(oct2016, 1))
    print()
    print("Oct 28 2020")
    print(round(oct2020, 1))
    
    ts.plot(df_2016, style=[".", ".", "--", "--", "-"], start_date="2016-07", color=2*party_colors)
    plt.title(state + " 2016");
    plt.show()
    ts.plot(df_2020, style=[".", ".", "--", "--", "-"], start_date="2020-07", color=2*party_colors)

    polling_error = 3.4
    lower_blue = df_2020.Biden_ave - polling_error
    upper_blue = df_2020.Biden_ave + polling_error
    lower_red = df_2020.Trump_ave - polling_error
    upper_red = df_2020.Trump_ave + polling_error

    plt.fill_between(df_2020.index, lower_blue, upper_blue, facecolor='dodgerblue', interpolate=True, alpha=0.2)
    plt.fill_between(df_2020.index, lower_red, upper_red, facecolor='red', interpolate=True, alpha=0.2)
    plt.xlim(("2020-07", df_2020.index[-1]))

    plt.title(state + " 2020");
    plt.show()
```
